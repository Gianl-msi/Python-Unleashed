```python
import numpy as np
import matplotlib.pyplot as plt
```

## OLS Visualized

```python
# True parameters of the model
intercept = 2  # intercept B0
b_real = 1 # slope B1

# Generate Data
X = np.random.normal(2, 30, size = 10)
y = b_real * X + intercept + np.random.normal(0, 18,  size = X.shape[0])

# Get X_min, X_max to set the limits of the plot   
x_int = np.array([np.min(X), np.max(X)])
```

```python
# plot the "observed" data
plt.figure(figsize = (6,6))
plt.scatter(X, y, s = 50, c = 'red', label = 'Observed')
plt.xlabel('Leg Length', fontsize =15)
plt.ylabel('Running Speed', fontsize =15)
plt.title('Observed Data', fontsize = 20)
plt.xticks(color='w')
plt.yticks(color='w')
plt.ylim([x_int[0]*1.5, x_int[1]*1.5])
plt.xlim([x_int[0]*1.5, x_int[1]*1.5])
plt.legend(loc = 'upper left', fontsize = 15)
plt.tight_layout()
plt.show()
```

## Guess the best fitting line - 1st guess

```python
# Initial guess for the fitting line 
b_initial = 0.2
intercept = 0

# Draw fitted line on guessed B0, B1
y_int = b_initial * x_int + intercept

# Calculate MSE
MSE = int(np.sum((y - (b_initial*X+ intercept))**2) / X.shape[0])
print(f"MSE: {MSE}")

# Plot observed data with guessed fitted line
plt.figure(figsize = (6,6))
plt.title('Observed Data + first line guess', fontsize = 20)
plt.scatter(X, y, s = 50, c = 'red', label = 'Observed')
plt.plot(x_int, y_int, c = 'blue', label = '1st guess')
plt.xlabel('Leg Length', fontsize =15)
plt.ylabel('Running Speed', fontsize =15)
plt.xticks(color='w')
plt.yticks(color='w')
plt.ylim([x_int[0]*1.5, x_int[1]*1.5])
plt.xlim([x_int[0]*1.5, x_int[1]*1.5])
plt.legend(loc = 'upper left', fontsize = 15)
plt.tight_layout()
plt.show()

# Plot observed data with guessed fitted line, and squared errors
plt.figure(figsize = (6,6))
plt.scatter(X, y, s = 80, c = 'red', label = 'Observed')
plt.plot(x_int, y_int, c = 'blue', label = f'1st guess, MSE: {MSE}')
plt.xlabel('Leg Length', fontsize =15)
plt.ylabel('Running Speed', fontsize =15)
plt.ylim([x_int[0]*1.5, x_int[1]*1.5])
plt.xlim([x_int[0]*1.5, x_int[1]*1.5])
plt.xticks(color='w')
plt.yticks(color='w')

# Add the squared residuals to the plot
delta_y = list(y - (b_initial*X + intercept))
yi_list = list(b_initial*X + intercept)
y_true = list(y)
Xs = list(X)

for num  in range(X.shape[0]):
    yi = yi_list[num]
    xi = Xs[num]
    plt.plot([xi, xi],[y_true[num], y_true[num] - delta_y[num]], c = 'orange')
    plt.plot([xi, xi - delta_y[num]],[y_true[num], y_true[num]], c = 'orange')
    plt.plot([xi, xi - delta_y[num]],[y_true[num]- delta_y[num], y_true[num]- delta_y[num]], c = 'orange')
    plt.plot([xi - delta_y[num], xi - delta_y[num]],[y_true[num] , y_true[num] - delta_y[num]], c = 'orange')

plt.title('Observed Data + line guess + square of residuals', fontsize = 20)
plt.scatter(Xs, yi_list, s = 80, label ='Predicted')   
plt.legend(loc = 'upper left', fontsize = 15)
plt.tight_layout()
plt.show()
```

## Estimate the best fitting line

### 1- Closed form formula 

The closed form formula to calculate W is  

```python
# We need to add a column of ones before the Xi in order to calculate the intercept B0
Xb0 = np.stack([np.ones(X.shape), X]).T
Xb0
```

```python
# Let's calculate the first half of the formula, (XTX)-1
XTX_1 = np.linalg.inv(np.dot(Xb0.T, Xb0))
XTX_1
```

```python
# Now the rest of the formula
b0_b1 = np.dot(XTX_1, np.dot(Xb0.T, y.reshape(-1,1)))

# Slope and intercept of the closed form solution: 
print('Intercept B0: ', b0_b1[0][0], '\nSlope B1: ', b0_b1[1][0])
```

**Note that the real values are: 2 for the intercept B0, and 1 for slope B1.**

```python
# Draw fitted line on estimated B0, B1
y_int = b0_b1[1][0] * x_int + b0_b1[0][0]

# Calculate final model MSE
MSE = int(np.sum((y - (b0_b1[1][0] * X + b0_b1[0][0]))**2) / X.shape[0])

# Plot observed data and fitted line
plt.figure(figsize = (8,8))
plt.plot(x_int, y_int, c = 'blue', label = f'Fitted line, MSE: {int(MSE)}')
plt.title('Observed Data with fitted line (closed form solution)', fontsize = 15)
plt.scatter(X,y, label = 'Observed Data', color ='red')
plt.ylim([x_int[0]*1.5, x_int[1]*1.5])
plt.xlim([x_int[0]*1.5, x_int[1]*1.5])
plt.legend(loc = 'upper left', fontsize = 15)
plt.xlabel('Leg Length', fontsize =15)
plt.ylabel('Running Speed', fontsize =15)
plt.xticks(color='w')
plt.yticks(color='w')
plt.tight_layout()
plt.show()
```

### 2- Gradient descent 

```python
# Function that calculates the loss
def Loss(y, y_hat): 
    return np.mean((y - y_hat)**2)
            
# Function that calculates gradients w.r.t. B1 and B0            
def calculate_gradients(y, y_hat, X):

    dL_dB0 =  2 * np.mean(y_hat - y)           # gradient w.r.t. B0, intercept
    dL_dB1 =  2 * np.mean((y_hat - y) *  X)    # gradient w.r.t. B1, slope
            
    return dL_dB0, dL_dB1  
```

```python
lr = 0.0005      # learning rate
n_iter = 15      # number of iterations

B0 = 0 # Inialize B0 with a 0  
B1 = 0 # Inialize B1 with a 0 

MSE = [] # Empty list to collect loss at each iteration

for i in range(n_iter):
    y_hat = X * B1 + B0   # making prediction 
    mse = Loss(y, y_hat)  # calculate loss
    MSE.append(mse)       # append loss
    
    dL_dB0, dL_dB1 = calculate_gradients(y, y_hat, X) # calculate the gradients
    
    B0 = B0 - dL_dB0 *  lr  # updating B0
    B1 = B1 - dL_dB1 *  lr  # updating B1


## PLOT RESULTS:
# Draw fitted line on estimated B0, B1
y_int = B1 * x_int + B0

# Plot loss vs iterations 
plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
plt.plot(np.arange(1, n_iter +1), MSE)
plt.title('Loss', fontsize = 15)
plt.xlabel('Iteration', fontsize = 12)

# Plot observed data and fitted line 
plt.subplot(1,2,2)
plt.plot(x_int, y_int, c = 'blue', label = f'Final Fitted line, MSE: {int(MSE[-1])}') # The last value is the final MSE
plt.title('Observed Data with fitted line (gradient descent)', fontsize = 15)
plt.scatter(X,y, label = 'Observed Data', color ='red')
plt.ylim([x_int[0]*1.5, x_int[1]*1.5])
plt.xlim([x_int[0]*1.5, x_int[1]*1.5])
plt.legend(loc = 'upper left', fontsize = 15)
plt.xlabel('Leg Length', fontsize =15)
plt.ylabel('Running Speed', fontsize =15)
plt.xticks(color='w')
plt.yticks(color='w')
plt.tight_layout()
plt.show()
```
